```python
import os
import logging
import argparse
from transformers import GPT2Tokenizer
import re
import sys

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def preprocess_line(line, tokenizer):
    # Clean and preprocess the line
    line = line.strip()
    # Remove unwanted characters using regex
    line = re.sub(r'[^A-Za-z0-9\s.,!?\'"-]', '', line)
    # Normalize spaces
    line = re.sub(r'\s+', ' ', line)
    # Tokenize the line
    tokenized_line = tokenizer.tokenize(line)
    # Decode tokens to handle special characters like ƒ†
    decoded_line = tokenizer.convert_tokens_to_string(tokenized_line)
    # Remove unwanted characters from the decoded line
    decoded_line = re.sub(r'[^A-Za-z0-9\s.,!?\'"-]', '', decoded_line)
    return decoded_line

def preprocess_data(input_file, output_file, tokenizer_name="gpt2"):
    try:
        tokenizer = GPT2Tokenizer.from_pretrained(tokenizer_name)
        with open(input_file, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        with open(output_file, 'w', encoding='utf-8') as f:
            for line in lines:
                if line.strip():  # Skip empty lines
                    processed_line = preprocess_line(line, tokenizer)
                    f.write(processed_line + "\n")

        logging.info(f"Preprocessing completed. Processed data saved to {output_file}")
    except FileNotFoundError:
        logging.error(f"File not found: {input_file}")
    except Exception as e:
        logging.error(f"An error occurred: {e}")

if __name__ == "__main__":
    # Check if running in an interactive environment
    if 'ipykernel' in sys.modules:
        input_file = "guvi_data.txt"  # Your company-specific data file
        output_file = "processed_guvi1_data.txt"
        preprocess_data(input_file, output_file)
    else:
        parser = argparse.ArgumentParser(description="Preprocess text data using GPT-2 tokenizer")
        parser.add_argument("input_file", type=str, help="Path to the input file containing raw text data")
        parser.add_argument("output_file", type=str, help="Path to the output file to save processed data")
        parser.add_argument("--tokenizer_name", type=str, default="gpt2", help="Name of the tokenizer to use (default: gpt2)")

        args = parser.parse_args()
        preprocess_data(args.input_file, args.output_file, args.tokenizer_name)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments, TextDataset, DataCollatorForLanguageModeling
from datasets import load_dataset, Dataset
import os

# Load pre-trained model and tokenizer
model_name = "gpt2"
tokenizer.pad_token = tokenizer.eos_token
# tokenizer = GPT2Tokenizer.from_pretrained(model_name)
# model = GPT2LMHeadModel.from_pretrained(model_name)

# Load and prepare dataset using datasets library
def load_and_prepare_datasets(file_path, tokenizer, block_size=128):
    # Load the text file into a dataset
    dataset = load_dataset('text', data_files={'train': file_path})

    # Tokenize the dataset
    def tokenize_function(examples):
        return tokenizer(examples['text'], return_special_tokens_mask=True)

    tokenized_datasets = dataset.map(
        tokenize_function,
        batched=True,
        remove_columns=["text"]
    )

    # Group texts into blocks
    def group_texts(examples):
        concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
        total_length = len(concatenated_examples[list(examples.keys())[0]])
        if total_length >= block_size:
            total_length = (total_length // block_size) * block_size
        result = {
            k: [t[i:i + block_size] for i in range(0, total_length, block_size)]
            for k, t in concatenated_examples.items()
        }
        return result

    tokenized_datasets = tokenized_datasets.map(group_texts, batched=True)

    return tokenized_datasets["train"]

# File path to your training text file
file_path = "D:/PythonProjects/GPTMODELUSINGHUGGINGFACE/guvi_data.txt"

# Prepare datasets
train_dataset = load_and_prepare_datasets(file_path, tokenizer)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=10,  # Reduce epochs for quicker experimentation
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    save_steps=10000,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=100,
    evaluation_strategy="steps",
    eval_steps=500,
    learning_rate=5e-5,
    weight_decay=0.01,
    warmup_steps=500,
    gradient_accumulation_steps=8,  # Accumulate gradients for better GPU utilization
    fp16=True,  # Use mixed precision if supported by your hardware
)

# Initialize data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=train_dataset,  # Use part of the train dataset for evaluation in this example
)

# Fine-tune the model
trainer.train()

# Save the fine-tuned model and tokenizer
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")
```

```python

from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the fine-tuned model and tokenizer
model_name_or_path = "./fine_tuned_model"
model = GPT2LMHeadModel.from_pretrained(model_name_or_path)
tokenizer = GPT2Tokenizer.from_pretrained(model_name_or_path)

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Define the text generation function
def generate_text(model, tokenizer, seed_text, max_length=100, temperature=1.0, num_return_sequences=1):
    # Tokenize the input text
    input_ids = tokenizer.encode(seed_text, return_tensors='pt').to(device)

    # Generate text
    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_length=max_length,
            temperature=temperature,
            num_return_sequences=num_return_sequences,
            do_sample=True,
            top_k=50,
            top_p=0.95,
        )

    # Decode the generated text
    generated_texts = []
    for i in range(num_return_sequences):
        # Adjust index if output shape requires it
        generated_text = tokenizer.decode(output[i], skip_special_tokens=True)
        generated_texts.append(generated_text)

    return generated_texts

# Test the model
seed_text = "Refund Policies"
generated_texts = generate_text(model, tokenizer, seed_text, max_length=100, temperature=0.7, num_return_sequences=3)

for i, text in enumerate(generated_texts):
    print(f"Generated Text {i + 1}:\n{text}\n")
```
